```python
# ============================================================
# PART 2: Practical Implementation (60%)
# Topic: AI in Software Engineering
# ============================================================

# ------------------------------
# üß† Task 1: AI-Powered Code Completion
# ------------------------------

# Manual Implementation: Sort list of dictionaries by a given key
data = [
    {"name": "Alice", "age": 25},
    {"name": "Bob", "age": 22},
    {"name": "Charlie", "age": 30}
]

def sort_by_key(data_list, key):
    """Sort a list of dictionaries by a specific key"""
    return sorted(data_list, key=lambda x: x[key])

sorted_data_manual = sort_by_key(data, "age")
print("Sorted Data (Manual):", sorted_data_manual)

# Copilot Suggested Version: More concise
data_copilot = [
    {"name": "Alice", "age": 25},
    {"name": "Bob", "age": 22},
    {"name": "Charlie", "age": 30}
]
sorted_data_copilot = sorted(data_copilot, key=lambda x: x["age"])
print("Sorted Data (Copilot):", sorted_data_copilot)




```

```python
# ------------------------------
# üß™ Task 2: Automated Testing with AI
# ------------------------------
# ‚ö†Ô∏è Note: Requires Selenium installed and ChromeDriver in PATH
# pip install selenium

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

def test_login_page():
    """Automate login testing for valid and invalid credentials"""
    driver = webdriver.Chrome()

    # URL of login page
    driver.get("https://example.com/login")  # Replace with your actual test URL
    time.sleep(2)

    # ‚úÖ Test valid credentials
    driver.find_element(By.NAME, "username").send_keys("test_user")
    driver.find_element(By.NAME, "password").send_keys("correct_password")
    driver.find_element(By.NAME, "password").send_keys(Keys.RETURN)
    time.sleep(2)

    if "dashboard" in driver.current_url.lower():
        print("‚úÖ Valid login successful")
    else:
        print("‚ùå Valid login failed")

    # ‚ùå Test invalid credentials
    driver.get("https://example.com/login")
    time.sleep(2)
    driver.find_element(By.NAME, "username").send_keys("wrong_user")
    driver.find_element(By.NAME, "password").send_keys("wrong_password")
    driver.find_element(By.NAME, "password").send_keys(Keys.RETURN)
    time.sleep(2)

    if "error" in driver.page_source.lower():
        print("‚úÖ Invalid login correctly rejected")
    else:
        print("‚ùå Invalid login test failed")

    driver.quit()

# Uncomment to run tests (ensure Selenium setup first)
test_login_page()
```

```python
# ------------------------------
# üß© Task 4: Code Refactoring with AI
# ------------------------------

# Original Code (Before Refactoring)
def calc_discount(price, discount):
    if discount > 0:
        new_price = price - (price * discount)
    else:
        new_price = price
    return new_price

# AI Refactored Version (Copilot Suggestion)
def apply_discount(price: float, discount: float = 0.0) -> float:
    """Apply a discount to the given price."""
    return price * (1 - discount) if discount > 0 else price

print("Original Function:", calc_discount(100, 0.1))
print("Refactored Function:", apply_discount(100, 0.1))

```

```python
# ------------------------------
# üß† Task 5: AI in Debugging Assistance
# ------------------------------

def buggy_sum(a, b):
    # Intentional bug: should be return a + b
    return a - b

def test_buggy_sum():
    assert buggy_sum(2, 3) == 5, "Bug detected in buggy_sum function"

try:
    test_buggy_sum()
except AssertionError as e:
    print("‚ùå Test failed:", e)
    print("üß© Copilot Suggestion: Replace '-' with '+' in buggy_sum")

# Corrected version
def fixed_sum(a, b):
    return a + b

print("‚úÖ Fixed sum result:", fixed_sum(2, 3))

# ============================================================
# END OF PART 2
# ============================================================
```
